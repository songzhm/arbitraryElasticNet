```python
# eg_dict = dict['eg_5']
def plott(eg_dict):
    mse=[eg_dict[key]['mse'] for key in eg_dict]
    se=[eg_dict[key]['se'] for key in eg_dict]
    plt.rcParams["figure.figsize"] = (7,5)
    colors=["steelblue"]*4+["sandybrown"]*5
    ax=plt.bar(range(len(mse)), mse,
            yerr=se, align='center',color=colors)
    rects = ax.patches
    labels = np.argsort(np.argsort(mse))+1
    for rect, label in zip(rects, labels):
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width() / 2, height+max(se), label,
                ha='center', va='bottom')
    plt.xticks(range(len(mse)), eg_dict.keys())
    plt.ylabel('Median mean-squared errors') 
    plt.ylim([0,7])
    custom_lines = [Line2D([0], [0], color="steelblue", lw=7),
                Line2D([0], [0], color="sandybrown", lw=7)]

    plt.legend(custom_lines, ['Common models', 'Generalized models'])
#     for i in [1,4]:
#         plt.xticks()[1][i].set_color("darkred")
#     for i in [2,5]:
#         plt.xticks()[1][i].set_color("darkolivegreen")
#     for i in [3,6,7,8]:
#         plt.xticks()[1][i].set_color("darkturquoise")
    plt.show()
```
